```python
import os
import pandas as pd
import datetime
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import sklearn.metrics as skm
from sklearn import preprocessing
from sklearn import linear_model
from sklearn.metrics import explained_variance_score,mean_absolute_error
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

%matplotlib inline
```

## Best Model -  XGB

In the next two steps, we are going to load our dataset and prepare it to be run on our model

```python
df = pd.read_csv('data/2_state_data.csv', index_col=0)
dfd = pd.get_dummies(df['State'])

for series in dfd:
    dfd[series] = dfd[series].astype(float)
dfd = pd.concat([df.drop(['State'],axis=1),dfd],axis=1)

X = dfd.drop(columns=['deaths_per_100k','deaths']).values
y = dfd.deaths_per_100k.values
```

```python
scaler = StandardScaler()
X = scaler.fit_transform(X)
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=123)
y_train = y_train.reshape(-1,1)
y_test = y_test.reshape(-1,1)
```

Previous exploration indicated that the best model was the XGB Regressor, so we will import the XGB regressor and run our model.

```python
import xgboost as xgb
from xgboost import XGBRegressor
from sklearn.metrics import r2_score, mean_squared_error, explained_variance_score

xgbreg = XGBRegressor(random_state=123,silent=True)
xgbreg.fit(X_train,y_train)
y_pred = xgbreg.predict(X_test)
```

```python
print("R2 Score: " + str(r2_score(y_test, y_pred)))
print("MSE Score: " + str(mean_squared_error(y_test,y_pred)))
print("Explained Variance Score: " + str(explained_variance_score(y_test,y_pred)))
```

```python
y_pred = y_pred.reshape(-1,1)
plt.scatter(x=y_test, y=y_pred)
plt.xlim(0,5)
plt.ylim(0,4)
plt.xlabel('Actual Values')
plt.ylabel('Model Predicted Values')
plt.show()
```

Now, we will print the feature importance to determine which factors have the most impact in driving the deaths/100k rate.

```python
features = pd.DataFrame(xgbreg.feature_importances_, xgbreg.get_booster().feature_names)
features['Name'] = dfd.drop(columns=['deaths_per_100k','deaths']).columns
relevant_features = features[features[0]!=0].sort_values(by=0, ascending=False)
relevant_features
```

```python
f, ax1 = plt.subplots(figsize=(12, 6))

x=relevant_features['Name']
y=relevant_features[0]
sns.barplot(x=x, y=y, palette="PuOr")
ax1.set_xticklabels(x, rotation=90)
ax1.set_xlabel('Weighted Factors')
ax1.set_ylabel('Weighted Importance')
plt.show()
plt.savefig('factors.png',bbox_inches='tight')
```
